# Business Problems
___

One year has passed since COVID-19 pandemic spread throughout the world, slowing down the economic growth in all parts of the world. Many countries applied lockdown as their measures to contain the spread of the virus, resulting in slump demand for tourism all over the world. But, recently some countries has lift the lockdown measures as their effort to suppressed the COVID-19 cases. Resulting in thriving economic post-pandemic and increasing demand for tourism.

Singapore is one of those successful countries to fight against pandemic. They start to open their border for travellers and hopefully can revive its tourism industry as soon as possible. Singapore’s tourism industry itself is vital for Singapore as it makes 4.1% of their national GDP (2017).

As the tourism industry started to grow, one of the important thing the traveler will need is accommodation as a place to spend the night.

Aside from hotel, Airbnb is another choice for traveller to look for accomodation in their travel destination. Airbnb is an American vacation rental online marketplace company. Through the service, users can arrange lodging, primarily homestays, and tourism experiences or list their properties for rental. Airbnb does not own any of the listed properties; instead, it profits by receiving commission from each booking.

As the lockdown measure has been lifted and tourism demand increases, Singapore resident can rent their properties in Airbnb to gain extra income sources and help to boost the economic growth in their countries. But, choosing the right competitive price is difficult since there are so many factors to be considered and calculated.

Machine learning can help to do those calculations and predict the correct listing price for new host as accurate as possible. This model can also be helpful for former host to revalue their listing into new competitive price and also for traveller to plan their budget for accommodations suiting their needs.

### Add New Features

In the listings dataset, there's `bathrooms_text` column which has the description about bathroom. But, we need to prerprocessed this column by generating new features in order for this information can be processed better by our model. Our plan is to create two more columns from this, that is number of bathrooms and bathroom type

```python
listings_clean.bathrooms_text.unique()
```

Some of the bathrooms text doesn't have number in it. So, we'll start by giving number first

```python
def add_number_bathrooms(value):
    if pd.isna(value):
        return np.nan
    else:
        if not any(map(str.isdigit, str(value))):
            return '1 ' + value
        else:
            return value
```

```python
listings_clean.bathrooms_text.apply(lambda x: add_number_bathrooms(x)).unique()
```

Next, we'll separate between number of bathrooms and type of bathrooms

```python
def separate_bathrooms(value):
    if not pd.isna(value):
        splitted = value.split()
        count = float(splitted[0])
        bath_type = ' '.join(splitted[1:]).lower()
        if bath_type[-1] == 's':
            bath_type = bath_type[:-1]
        return count, bath_type
    else:
        return np.nan
```

```python
listings_clean.bathrooms_text.apply(lambda x: add_number_bathrooms(x)).apply(lambda x: separate_bathrooms(x))
```

Then, we'll assign it to different columns and delete the original columns

```python
listings_clean['bathrooms_text'] = (listings_clean.bathrooms_text
                                    .apply(lambda x: add_number_bathrooms(x))
                                    .apply(lambda x: separate_bathrooms(x)))
```

```python
listings_clean['total_bathrooms'] = listings_clean.bathrooms_text.apply(lambda x: x[0] if not pd.isna(x) else np.nan)
listings_clean['bathrooms_type'] = listings_clean.bathrooms_text.apply(lambda x: x[1] if not pd.isna(x) else np.nan)
```

The table above shows each columns that we will use to build our model for predicting price of Singapore's Airbnb Listing along with its data type and its unique value. We need to encode some of these columns in order to build machine learning model. Thus, the encoding strategy will be:
- For columns with boolean value (True/False), we will use One-Hot Encoder and map it to value of 0 and 1
- For categorical columns, we will use One-Hot Encoder for columns with number of unique value less than 10. Otherwise, we will use Binary Encoder to minimize the number of columns
- For numerical columns, we will leave it as it is

Several things to note:

- Our data contains a lot of categorical and discrete values which we may take into account when choosing model for prediction

## Save to Database

```python
def chunker(seq, size):
    # from http://stackoverflow.com/a/434328
    return (seq[pos:pos + size] for pos in range(0, len(seq), size))

def insert_with_progress(df, tablename, if_exists="replace"):
    engine = db.create_engine(os.getenv('db-uri'))
    with engine.connect() as con:
        chunksize = int(len(df) / 10) # 1%
        with tqdm(total=len(df)) as pbar:
            for i, cdf in enumerate(chunker(df, chunksize)):
                replace = if_exists if i == 0 else "append"
                cdf.to_sql(con=con, name=tablename, if_exists=replace, index=False)
                pbar.update(chunksize)
```

```python
insert_with_progress(calendar_clean, 'calendar')
insert_with_progress(reviews_clean, 'reviews')
insert_with_progress(neighbourhood_clean, 'neighbourhood')
insert_with_progress(listings_clean, 'listings')
```
