## Calendar Dataset

```python
calendar.head()
```

```python
calendar.dtypes
```

```python
calendar_clean1 = calendar.copy()
calendar_clean1.sample(5)
```

### Change Incorrect Data Types

#### `date`

The `date` column represents the respective data for each record. Currently, the data type is `object` while it should be in `datetime`

```python
calendar_clean1['date'] = pd.to_datetime(calendar_clean1['date'], format='%Y-%m-%d')
calendar_clean1['date'].head()
```

#### `available`

`available` column represents the availability of the listing on each day. There are 2 unique values for `available` column, 't' and 'f' which stands for True and False

```python
calendar['available'].unique()
```

```python
calendar_clean1['available'] = np.where(calendar_clean1['available'] == 't', True, False)
calendar_clean1['available'].head()
```

#### `price` and `adjusted_price`

Column `price` and `adjusted_price` have object data type of `object` whereas it should have been numeric data type, such as float or int. Before changing it to numeric data types, we need to get rid of dollar sign in those columns and determine whether those columns can be casted into float or int

```python
# Make sure that all records started with dollar sign
(calendar['price'].str[0] != '$').sum()
```

```python
calendar_clean1['price'] = calendar_clean1['price'].apply(lambda x: x[1:])
calendar_clean1['price'].head()
```

```python
calendar_clean1['adjusted_price'] = calendar_clean1['adjusted_price'].apply(lambda x: x[1:])
calendar_clean1['adjusted_price'].head()
```

We also need to delete comma as thousands separator in these columns

```python
calendar_clean1[calendar_clean1['price'].apply(lambda x: ',' in x)]
```

```python
calendar_clean1['price'] = calendar_clean1['price'].str.replace(',','')
```

```python
calendar_clean1[calendar_clean1['adjusted_price'].apply(lambda x: ',' in x)]
```

```python
calendar_clean1['adjusted_price'] = calendar_clean1['adjusted_price'].str.replace(',','')
```

Next, we need to check if we should convert it to float or int. The difference is whether there's price with fraction in it. If it is, then we need to convert it as float and vice versa

```python
# Check if there's any price that should be converted as float
(calendar_clean1['price'].astype(float) % 1 != 0).sum()
```

```python
# Check if there's any adjusted_price that should be converted as float
(calendar_clean1['adjusted_price'].astype(float) % 1 != 0).sum()
```

With little checking, we can safely convert `price` and `adjusted_price` column into int

```python
calendar_clean1['price'] = calendar_clean1['price'].apply(lambda x: int(x.split('.')[0]))
calendar_clean1['price'].head()
```

```python
calendar_clean1['adjusted_price'] = calendar_clean1['adjusted_price'].apply(lambda x: int(x.split('.')[0]))
calendar_clean1['adjusted_price'].head()
```

#### `minimum_nights` and `maximum_nights`

The `minimum_nights` and `maximum_nights` column represent minimum and maximum number of nights possible for user to book in each day for each listing. Normally, it should be int data type. But, there's some null values so we'll leave the data type as it is

```python
calendar_clean1['minimum_nights'].isnull().sum()
```

```python
calendar_clean1['maximum_nights'].isnull().sum()
```

#### Other Columns

Other columns already has correct data type, so we'll leave it as it is

### Drop Unnecessary Columns

For the calendar dataset, all columns may will be used for further analysis. Therefore, we will not any columns in this dataset

## Listings Dataset

```python
listings.head(2)
```

```python
listings.dtypes
```

```python
listings_clean1 = listings.copy()
listings_clean1.head()
```

### Change Incorrect Data Types

#### `price`

The `price` column have object data type where it should be numerical data type, either float or int. To convert it as numerical data type, there are several things we need to do first

First, we need to remove the dollar sign on each price

```python
# Check if each price is started with dollar sign
listings_clean1['price'].apply(lambda x: x[0] != '$').sum()
```

```python
listings_clean1['price'] = listings_clean1['price'].str[1:]
listings_clean1['price'].head()
```

Then, we also need to delete comma as thousand separator before casting it to numerical data type

```python
# Check if there's price with comma as thousand separator
listings_clean1[listings_clean1['price'].apply(lambda x: ',' in x)]['price']
```

```python
listings_clean1['price'] = listings_clean1['price'].str.replace(',','')
```

Lastly, we need to figure out whether we should convert it as int or float by checking if there's fraction in price

```python
# Check if any fraction in price. If there's, we need to convert it as float and vice-versa
(listings_clean1['price'].astype(float) % 1 != 0).sum()
```

Since all price is whole number, we can convert it as int instead of float

```python
listings_clean1['price'] = listings_clean1['price'].apply(lambda x: int(x.split('.')[0]))
listings_clean1['price'].head()
```

#### `last_scraped`, `host_since`, `first_review`, `last_review` and `calendar_last_scraped`

These four columns represents date time object meanwhile it has object datatype

```python
datetime_col = ['last_scraped', 'host_since', 'first_review', 'last_review', 'calendar_last_scraped']
```

```python
listings_clean1[datetime_col].head()
```

```python
listings_clean1[datetime_col] = listings_clean1[datetime_col].apply(pd.to_datetime, format='%Y-%m-%d')
listings_clean1[datetime_col].head()
```

#### `host_is_superhost`, `host_has_profile_pic`, `host_identity_verified`, `has_availability`, `instant_bookable`

```python
# Check the unique value of each column
listings_clean1[bool_col].apply(pd.unique)
```

```python
listings_clean1[bool_col].isnull().sum()
```

Turns out, we have several column with null values. We'll leave as it is for now and impute the null values later. For now, we will only convert non-null values to boolean 

```python
def convert_bool_col(ser):
    if pd.isna(value):
        return np.nan
    else:
        return value == 't'
```

```python
listings_clean1[bool_col] = listings_clean1[bool_col].applymap(convert_bool_col)
listings_clean1[bool_col].head()
```

#### `host_response_rate`, `host_acceptance_rate`

These columns represent percentage of host response and host acceptance of order in their listings. It should be converted as numerical before further analysis

```python
numeric_col1 = ['host_response_rate', 'host_acceptance_rate']
```

```python
listings_clean1[numeric_col1]
```

We need to remove the percentage on each value also leave null values as it is. Then, we also need to convert it to numerica data types. In the end, the data type for these columns will be float since it has null values

```python
# Check whether every non-null values is ended with percentage sign
listings_clean1[numeric_col1].applymap(lambda x: x[-1] != '%' if not pd.isna(x) else False).sum()
```

```python
listings_clean1[numeric_col1] = listings_clean1[numeric_col1]\
                                .applymap(lambda x: int(x.replace('%','')) if not pd.isna(x) else np.nan)
listings_clean1[numeric_col1].head()
```

Other columns already has correct data type, so we'll leave it as it is

From result above, we know that column `license`, `calendar_updated`, `bathrooms` has all null values for every record. Therefore, we will drop these three columns

```python
listings_clean1 = listings_clean1.drop(['license', 'calendar_updated', 'bathrooms'], axis=1)
```
